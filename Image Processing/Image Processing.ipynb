# image Processing

```python
from PIL import Image
```

```python
filename = "buildings.jpg"
with Image.open(filename) as img:
    img.load()
    
```

```python
type(img)
```

```python
isinstance(img, Image.Image)
```

```python
img.show
```

```python
cropped_img = img.crop((300, 150, 700, 1000))
cropped_img.size
cropped_img.show()
low_res_img = cropped_img.resize(
    (cropped_img.width // 4, cropped_img.height // 4)
)
low_res_img.show()
```

```python
low_res_img = cropped_img.reduce(4)
```

```python
cropped_img.save("cropped_image.jpg")
low_res_img.save("low_resolution_cropped_image.png")
```

```python
converted_img = img.transpose(Image.FLIP_TOP_BOTTOM)
converted_img.show()
```

```python
rotated_img = img.rotate(45)
rotated_img.show()
```

```python
rotated_img = img.rotate(6, expand=True)
rotated_img.show()
```

```python
#Bands and Modes of an Image
filename = "strawberry.jpg"
with Image.open(filename) as img:
    img.load()
```

```python
cmyk_img = img.convert("CMYK")
gray_img = img.convert("L")  # Grayscale
```

```python
cmyk_img.show()
gray_img.show()
```

```python
img.getbands()
```

```python
cmyk_img.getbands()
```

```python
gray_img.getbands()
```

```python
red, green, blue = img.split()
red
```

```python
red.mode
```

```python
 #create three new RGB images showing the red, green, and blue channels
zeroed_band = red.point(lambda _: 0)
red_merge = Image.merge(
    "RGB", (red, zeroed_band, zeroed_band)
)
green_merge = Image.merge(
    "RGB", (zeroed_band, green, zeroed_band)
)
blue_merge = Image.merge(
    "RGB", (zeroed_band, zeroed_band, red)
)
red_merge.show()
green_merge.show()
blue_merge.show()
    
```
